```python
import pandas as pd
import numpy as np
```

```python
d={}
studentName=[]
studentAge=[]
studentNumber=[]
studentLesson=[]
studentGrade =[]
global df 
df= pd.DataFrame(d)
df=df.assign(StudentName = studentName,StudentAge=studentAge, StudentNumber=studentNumber,StudentLesson=studentLesson,StudentGrade=studentGrade)
```

```python
def addStudent(df):
  name = input("İsim giriniz : ")
  age = input("Yaş giriniz : ")
  isInteger = True
  while(isInteger):
      try:
          age = int(age)
          isInteger = False
      except ValueError:
          age = input("Lütfen Sayı Türünde Bir Değer Giriniz \nYaş giriniz : ")
      
  #while type(int(yas))==str:
  #  yas = int(input("Lütfen Sayı Türünde Bir Değer Giriniz \nYaş giriniz : "))
 
  studentNo =int(input("Okul Numaranızı giriniz : "))
  while 1000 > studentNo or studentNo > 5000:
        studentNo =int(input("Lütfen 1000-5000 Arasında Bir Numara Girişi Yapınız \nOkul Numaranızı giriniz : "))
  if (df['StudentNumber']==studentNo).any() :
    print("Mevcut Bir Numara Girişi Yapılmaya Çalıştın! Lütfen Tekrar Numara Giriniz! ")
    studentNo =int(input("Okul Numaranızı giriniz : "))
  
  lessonName="Linear Algebra"
  
  studentScore = int(input("Notunuzu giriniz: "))
  while 0 > studentScore or studentScore > 100 :
    studentScore = int(input("Lütfen 0-100 Arasında Bir Not Girişi Yapınız \nNotunuzu giriniz: ")) 

  df=transferInfo(df,name,age,studentNo,lessonName,studentScore)
  return df
```

```python
def transferInfo(df,name,age,studentNo,lessonName,studentScore):
  studentName.append(name)
  studentAge.append(age)
  studentNumber.append(studentNo)
  studentLesson.append(lessonName)
  studentGrade.append(studentScore)
  ogr = {
      "StudentName": [name],
      "StudentAge": [age],
      "StudentNumber":[studentNo],
      "StudentLesson":[lessonName],
      "StudentGrade":[studentScore]
    }
  data = pd.DataFrame(ogr)
  df = df.append(data)
  return df
```

```python
def menu(df):
  status=True
  while status:
    print("""
    1) Öğrenci Ekle
    2) Öğrenci Sorgula
    3) Bütün Öğrencileri Göster
    4) Öğrencilerin Harf Notunu Göster
    5) Öğrencilerin Kaldı/Geçti Durumunu Sorgula
    6) En Yüksek Nota Sahip Öğrenciyi Ekrana Yazdır
    7) En Düşük Nota Sahip Öğrenciyi Ekrana Yazdır
    8) Sınıfın Not Ortalamasını Yazdır
    9) Öğrencileri Alfabetik Sırala
    10) Verileri Excele Aktar
    11) Uygulamayı Kapat\n""")
    chosen = int(input("Seçim yapınız : "))

    if chosen == 1:
     df=addStudent(df)
     print("\nÖğrenci Başarıyla Eklendi.\n")
    if chosen == 2:
      studentNo=int(input("Öğrenci Numarasını Giriniz : "))
      print(df[df['StudentNumber'] == studentNo])
    if chosen == 3:
      print("\nÖğrenci Listeniz Başarıyla Ekrana Getirildi.\n")
      print(df)
    if chosen == 4:
      df['Harf'] = np.where((df.StudentGrade < 60 ), 
            'F', np.where((df.StudentGrade >= 60) & (df.StudentGrade <= 69),  
            'D', np.where((df.StudentGrade >= 70) & (df.StudentGrade <= 79), 
            'C', np.where((df.StudentGrade >= 80) & (df.StudentGrade <= 89), 
            'B', np.where((df.StudentGrade >= 90) & (df.StudentGrade <= 100), 
            'A', 'No Marks')))))
      print("\nHarf Notları Başarıyla Güncellendi\n")
      print(df)
    if chosen == 5:
      df['Kaldimi'] = np.where((df.Harf == 'F' ), 
                  'Kaldi', np.where((df.Harf != 'F'),  
                  'Geçti', 'No Marks'))
      print(df)
      print("\nKalan/Geçen Listesi Başarıyla Güncellendi.\n")
    if chosen == 6 :
      print(f"\nEn Yüksek Nota Sahip Öğrenci\n : {df[df.StudentGrade.max() == df['StudentGrade']]}")
    if chosen == 7 :
      print(f"\nEn Düşük Nota Sahip Öğrenci\n : {df[df.StudentGrade.min() == df['StudentGrade']]}")
    if chosen == 8:
      meanTotal= df["StudentGrade"].mean()
      print(f"\nSınıfın Not Ortalaması : {meanTotal}")
    if chosen == 9:
      print(df.sort_values(["StudentName"], ascending=True))
    if chosen == 10:
      df.to_excel("studentDF.xlsx",
             sheet_name='Sheet_name_1')
      print("Veriler Başarıyla Excele Aktarıldı")  
    if chosen == 11:
      print("\nProgram Kapatılıyor.....\n")
      status=False
```
